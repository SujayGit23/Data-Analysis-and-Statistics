### What is Multivariate outlier detection ?

```python
''' Use multivariate methods to find the outliers that only show up within combinations of observations
from two or more different variables '''
```

### What are the methods for multivariate outlier detection ?

```python
'''
- Scatterplot matrices
- Boxplots '''
```
