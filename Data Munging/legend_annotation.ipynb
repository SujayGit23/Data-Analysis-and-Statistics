```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
cars = pd.read_csv("D:\Workspace\Data Munging\mtcars.csv")
print(cars.head())
```

```python
rcParams['figure.figsize'] = 8,4
sb.set_style('whitegrid')

x = range(1,10)
y = (1,2,3,4,0.5,4,3,2,1)
plt.bar(x,y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
plt.show()
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle','motorbike','car','van','stoller']
plt.pie(z,labels=veh_type)
plt.show()
```

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))

ax.set_xticklabels(cars.cars_name,rotation=60,fontsize='medium')
ax.set_title('Miles per gallon of cars in mtcars')
ax.set_xlabel('cars names')
ax.set_ylabel('miles/gallon')
ax.legend(loc='best')
plt.show()
```
