```python
'''
Two methods for Plot Building
1. Functional methods by calling the plotting function on a
variable or set of variables
2. Object oriented method - building a plot by generating a
blank fig object and then populating that object with plot and plot elements
'''
```

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
cars = pd.read_csv("D:\Workspace\Data Munging\mtcars.csv")
print(cars.head())

rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
plt.show()
cars.columns = ['car_names','mpg','cyl','disp','hp',
                'drat','wt','qsec','vs','am','gear','carb']
mpg = cars['mpg']

mpg.plot()
plt.show()
```

```python
df = cars[['cyl','wt','mpg']]
df.plot()
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
mpg.plot(kind='bar')
plt.show()
```

```python
mpg.plot(kind = 'barh')
plt.show()
```

```python
x = [1,2,3,4,0.5]
plt.pie(x)
plt.show()
```
